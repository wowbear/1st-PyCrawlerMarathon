```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(r.text, 'html.parser')
table = soup.find('table', class_='TABLE_G')
print(table)
```

```python
import re

d = {}
tbl = table.find_all('td')

for tag in tbl:
    if 'style' in tag.attrs:
        item = tag.text
        d[item] = {}
    if '/' in tag.text:
        yymm = tag.text            
    if re.match('^[-+]?[-0-9]\d*\.\d*|[-+]?\.?[0-9]\d*$',tag.text):
        d[item][yymm] = tag.text
        
print(d)
```
