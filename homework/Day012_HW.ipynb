### ① 取出今天所有的發文

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text, "html5lib")

for d in soup.find(class_="part_list_2").find_all('h3'):
    print(d.find(class_="date").text, d.find_all('a')[-1].text)

```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
dict_news = dict()

for d in soup.find(class_="part_list_2").find_all('h3'):
    #for item in d.find_all('em'):
    #    print(item.label.get_text())
    news_type = d.find('em').get_text()
    if news_type in dict_news:
        dict_news[news_type] += 1
    else:
        dict_news[news_type] = 1
        
print(dict_news)
```

### ③ 哪一個類別的文章最多

```python
print(sorted(dict_news.items(), key=lambda d: d[1])[-1])
```
